
This notebook implements and compares Q-learning and SARSA in an 11×11 Grid-World environment, then explores Interactive Reinforcement Learning via a Teacher-Student framework.


Hyper-parameters (identical across all tasks)
- learning_rate = 0.1  
- discount = 0.95  
- epsilon_start = 1.0 → epsilon_end = 0.05 (exponential decay)  
- episodes = 450  
- max_steps_per_episode = 100  



Evaluation Metrics
For every run we compute  
- Average Reward per Episode  
- Success Rate (%) 
- Average Steps to Goal 



Teacher-Student Framework
- Availability ∈ {0.1, 0.3, 0.5, 0.7, 1.0}  
- Accuracy   ∈ {0.1, 0.3, 0.5, 0.7, 1.0}  
- Teacher advises the greedy action (w.r.t. its frozen Q-table) with prob(availability)  
- If advising, the advice is correct with prob(accuracy); otherwise a random non-optimal action is returned.

-  Heat-maps (availability × accuracy)  
-  Teacher-impact curves (fixed availability = 0.5)  
-  Teacher-effectiveness summary (best teacher vs baseline)
